    - 주요특징

        - 다양한 그래프 지원: 선 그래프, 막대 그래프, 산점도, 히스토그램, 박스 그래프 등 기본적인 그래프 제작 가능

        - 대규모 데이터 처리: 다양한 데이터셋을 비교하고 세밀하게 조정

```python
import matplotlib.font_manager as fm
import os

font_dir = "/usr/share/fonts/truetype/nanum"

for root, dirs, files in os.walk(font_dir):
    for file in files:
        if file.endswith(".ttf"):
            full_path = os.path.join(root, file)
            fm.fontManager.addfont(full_path)

# 다시 로드된 폰트 확인
[x.name for x in fm.fontManager.ttflist if "Nanum" in x.name]
```

- Seaborn

        - Matplotlib과의 비교
            - Matplotlib은 기초부터 세밀하게 조정가능한 강력한 도구, Seaborn은 Matplotlib의 기능을 간단하고 직관적으로 사용할 수 있도록 설계된 상위 라이브러리
            - Seaborn은 Matplotlib을 내부적으로 사용하므로, 두 라이브러리를 함께 활용하면 더욱 유용

            - 통계적 시작화: 히트맵, 상관관계 매트릭스, 카테고리별 데이터 비교 등 통계적 데이터 시각화에 강점
            - 스타일 조정: 기본 제공되는 그래프 스타일이 세련되고 깔끔하여 별도의 커스터마이징 없이 바로 사용 가능
        - 활용 사례
            - 데이터 탐색: 데이터프레임을 활용한 분포 확인, 상관관계 분석

```python
import seaborn as sns
import pandas as pd

# 샘플 데이터프레임 생성
data = pd.DataFrame({
    "Category": ["A", "B", "C", "D"],
    "Values": [5, 7, 3, 8]
})

# 막대 그래프
sns.barplot(x="Category", y="Values", data=data)
plt.title("Bar Plot Example")
plt.show()
```

- import seaborn as sns
    - Seaborn은 파이썬의 고급 시각화 라이브러리로, 통계적 데이터를 시각화 하는데 적합, 여기서는 Seaborn을 sns라는 별칭 지정
- import pandas as pd

    - 데이터를 표 형태로 다룰수 있도록 지원하며, Seaborn과 함께 사용
- pd.DataFrame은 표 형식으로 데이터를 저장, 데이터느 열(Column)과 행(Row)으로 구성, Excel 표와 유사한 구조
- 데이터 구성
    - "Category": 각 막대의 이름을 지정하는 열, 여기서는 "A", "B", "C", "D"로 구성
    - "Values": 각 카테고리에 해당하는 값으로, 막대 그래프의 높이를 결정, 값은 5, 7, 3, 8
    - "sns.barplot": Seaborn의 barplot 함수는 카테고리 데이터를 기반으로 막대 그래프를 생성
- 매개변수 설명
    - x="Category": x축에 표시될 데이터의 열을 지정, 여기서는 "Category" 열이 x축에 표시, 값은 "A", "B", "C", "D"
    - y="Values": y축에 표시될 데이터 열을 지정, 여기서는 "Values" 열이 Y축 값으로 사용, 값은 5, 7, 3, 8 로 나타남
- data=data: 시각화에 사용할 데이터프레임을 지정, 위에서 생성한 data를 사용
- 결과
    - 각 카테고리("A", "B", "C", "D")에 대해 값(5, 7, 3, 8)이 Y축 높이로 설정된 막대 그래프가 생성

- plt.show()

- Plotly
    - Plotly는 대화형 시각화를 지원하는 고급 라이브러리로 웹 애플리케이션에서도 활용 가능
    - 정적 그래프를 넘어 동적이고 직관적인 시각화를 구현할 수 있는 도구
    - 주요특징
        - 대화형 그래프: 줌인, 줌아웃, 항목 강조 등과 같은 상호작용 기능 지원
        - 웹 통합: 생성한 그래프를 HTML로 내보낼 수 있어 웹 기반 프로젝트에 활용 가능
        - 3D 사각화: 3D 산점도, 지형도, 표면 플롯 등 고급 시각화 가능
    - 활용사례
        - 데이터 대화형 탐색: 사용자가 직접 데이터를 탐색하여 패턴을 발견할 수 있도록 지원
        - 웹 애플리케이션: 웹 프로젝트에서 데이터 시각화 모듈로 활용
        - 고급 분석: 3D 플롯을 통해 데이터의 구조와 관계를 다차원적으로 분석

```python
import plotly.express as px

# 샘플 데이터
data = px.data.gapminder()

# 대화형 산점도
fig = px.scatter(
    data,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="continent",
    log_x=True,
    title="Interrective Scatter Plot"
)

# 그래프 크기 조정
fig.update_layout(width=800, height=500)
fig.show()
```

- x="gdpPercap"
    - x축에 사용할 열을 지정, 여기서는 gdpPercap(1인당 GDP)을 x축에 설정하여 국가별 경제력 표시
- y="lifeExp"
    - y측에 사용할 열을 지정, 여기서는 lifeExp(기대수명)를 y축에 설정, 국가별 평균 기대수명을 나타냄
- size="pop"
    - 점(버블)의 크기를 나타낼 열을 지정, 여기서는 pop(인구)를 사용하여 각 국가의 점 크기를 인구에 비례하도록 설정
- color="continent"
    - 점의 색상을 결정할 열을 지정, 여기서응 continent(대륙) 정보를 사용하여 대륙별로 다른 색상 부여
    - asia: 아시아
    - Europe: 유럽
    - Africa: 아프리카
    - Americas: 아메리카
    - Oceania: 오세아니아
- log_x=True
    - x축의 값을 로그 스케일(log scale)로 변환, 로그 스케일은 데이터가 매우 큰 범위로 분포되어 있을 때, 이를 압축하여 시각적으로 보기 쉽게 만듬, 여기서는 GDP 데이터가 폭넓게 분포되어 있기 때문에 보기 쉬움
- title="Interractive Scatter Plot"
    - 그래프의 제목을 지정, 제목은 그래프 상단에 표시, 데이터의 요약 또는 그래프의 목적을 설명
- fig.show()
    - 그래프를 화면에 표시, Plotly는 대화형 그래프를 제공, 마우스를 이용해 그래프를 확대, 축소하거나 세부 정보를 탐색 가능
- fig.update_layout()
    - 그래프의 전반적인 레이아웃을 조정하는 함수
- width=800
    - 그래프의 너비를 픽셀 단위로 설정, 여기서는 800픽셀로 설정
- height
    - 그래프의 높이를 픽셀 단위로 설정, 여기서는 500픽셀로 설정

- Pandas Plot

    - 주요 특징

        - 빠른 결과 확인: 데이터 탐색 과정에서 시각화를 통해 빠르게 데이터를 분석

- 그래프 그리기 기본 코드 프레임워크 정리
    - 데이터를 시각화한다는 것은 숫자로만 되어 이는 데이터를 사람이 눈으로 이해하기 쉽게 그림으로 표현 한다는 뜻, 예를 들어 나이에 따른 콜레스테롤 변화, 성별에 따른 최대 심박수 차이 등을 그래프로 보면 패턴이나 차이가 훨씬 직관적으로 보임
    - 파이썬에서 가장 많이 사용하는 시각화 도구는 아래 두가지
        - matplotlib.pyplot:  기본 그래프 도구(단순하지만 강력함)
        - setborn: matplotlib을 기반으로 한 고극 시각화 도구 (예쁨 + 통계적 기능 내장)
    
    - 시각화 기본 흐름 (공통 구조)
        - 어떤 그래프를 그리든 기본적으로 생각할 흐름은 다음과 같음
            - x축 변수: 기준이 되는 항목 -> 연속형 데이터나 범주형 데이터
            - y축 변수: 수치형 결과 -> 평균, 합계, 개수 등
            - 그래프에 형태 선택: 선그래프, 막대그래프, 산점도
            - 그래프에 제목, 축 이름 붙이기
            - plt.show()로 그래프 출력

- matplotlib.pyplot: 가장 기본적인 기각화 도구
    - matplotlib.pyplot은 파이썬의 기본 그래프 라이브러리, 단순한 선, 막대, 산점도를 그릴 때 사용, 그래프의 각 요소를 세밀하게 설정할 수 있는 장점

```python
# matplotlib 라이브러리 및 한글폰트 출력, 경고메시지 중지 코드

# 시각화 라이브러리 임포트
# 시각화 출력 시 경고메시지 안나오게 처리
import matplotlib.pylab as plt
import matplotlib.font_manager as fm
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
 
# 한글 폰트 설정 (나눔 고딕(NanumGothic))
plt.rcParams["font.family"] = "NanumGothic"

# 마이너스 부호 깨짐 방지
plt.rcParams["axes.unicode_minus"] = False

# 선 그래프: 시간 흐름이나 연속적인 데이터 변화

# 데이터셋 임포트
df = pd.read_csv("datasets/heart.csv")

# 1. 나이별 평균 콜레스테롤 계산
age_chol = df.groupby("age")["chol"].mean().reset_index()

# 2. 선 그래프 그리기
plt.plot(age_chol["age"], age_chol["chol"])
plt.xlabel("나이")
plt.ylabel("평균 콜레스테롤 수치")
plt.title("나이별 평균 콜레스테롤 변화")
plt.grid(True)
plt.show()
# plt.plot(x, y) 는 선을 그리는 기본 구조
```

```python
# 막대 그래프: 항목별 수치 비교

# x축: 성별(0, 1), y축: 평균 콜레스테롤
plt.bar(df["sex"], df["chol"])
plt.xlabel("성별")
plt.ylabel("콜레스테롤 수치")
plt.title("성별에 다른 콜레스테롤 비교")
plt.show()
```

```python
# 산점도 그래프: 두 연속형 변수 간 관계 보기

# x축: 나이, y축: 최대 심박수
plt.scatter(df["age"], df["thalachh"])
plt.xlabel("나이")
plt.ylabel("최대 심박수")
plt.title("나이와 심박수의 관계")
plt.show()
```

- plt 구조 이래 4단계
    - 1단계: 틀 만들기 (그래프 종류)
        - plt.plot(x, y)        # 선 그래프
        - plt.bar(x, y)         # 막대 그래프
        - plt.scatter(x, y)     # 산점도
            - x와 y는 리스트나 Series 형식으로 넣으면 됨
            - 여기서부터 시작, 어떤 그래프를 지정할 지 고르는 단계

    - 2단계: 데이터 넣기
        - plt.plot(df["age"], df["chol"])
            - 위의 x, y에 원하는 데이터를 넣음
            - 보통 df["칼럼명"] 형태로 Series를 지정
    
    - 3단계: 이름 붙이기 (label)
        - plt.xlabel("x축 이름"), ex> 나이
        - plt.ylabel("y축 이름"), ex> 콜레스테롤
        - plt.title("그래프 제목"), ex> 나이별 콜레스테롤
            - 그래프를 보는 사람이 이해할 수 있도록 레이블을 붙이는 단계
            - 안 붙이면 그래프 파악이 어려움
    
    - 4단계: 보여주기
        - plt.show()
            - 이 부분이 작성되면 지정한 그래프가 표시됨

- 히스토그램(histogram)
    - 히스토그램이란?
        - 히스토그램은 데이터의 분포를 시각화 하는 데 사용, 데이터를 구간(bin)으로 나누어 각 구간에 속하는 데이터의 빈도를 보여줌, 데이터의 분포 상태를 이해하거나 이상치를 확인하는 데 유용
            - 구성 요소
                - x축: 데이터의 구간을 나타냄
                - y축: 각 구간에 속하는 데이터의 빈도를 나타냄
                - 막대: 데이터가 해당 구간에 얼마나 포함되어 있는 지를 표현
            - 활용 사례
                - 학생들의 시험 점수 분포 분석
                - 특정 제품의 일일 판매량 분포 파악
                - 센서 데이터를 통해 특정 값이 자주 발생하는 지 확인

```python
# 히스토그램은 데이터의 분포를 이해하거나, 특정 구간에 데이터가 몰려있는지를 확인할 때 유용, 
# 예를 들어, 특정 도시의 하루 평균 온도를 시각화하면, 대다수의 날씨기 온화한지 또는
# 극단적으로 더운 날씨가 잦은 지 알 수 있음

import pandas as pd
import matplotlib.pylab as plt

# 데이터 불러오기
data = pd.read_csv("datasets/heart.csv")

# 히스토그램 생성
plt.hist(
    data["age"],        # 나이 데이터를 시각화
    bins=10,            # 구간 수
    color="skyblue",    # 막대 색상
    alpha=0.7,          # 투명도
    edgecolor="black"   # 테두리 색상 
)

plt.title("Age Distribution in Heart Dataset")  # 그래프 제목
plt.xlabel("Age")                               # x축 이름
plt.ylabel("Frequency")                         # y축 이름
plt.grid(True)                                  # 격자 표시
plt.show()
```

- import pandas as pd
    - CSV 데이터를 읽고 처리하는 데 사용되는 pandas 라이브러리를 호출
- import matplotlib.pyplot as plt
    - 데이터를 시각화하기 위한 Matplotlib의 Pyplot 모듈을 호출
- data = pd.read_csv("datasets/heart.csv")
    - datasets/heart./csv 파일을 읽어 데이터프레임으로 변환, data 변수에 데이터프레임을 저장하여 나중에 분석 및 시각화에 사용, 파일 경로가 정확해야 정상적으로 로드 가능
- plt.hist(): 히스토그램을 생성하는 함수
    - data["age"]: heart.csv의 age 열 데이터를 사용
    - bins=10: 데이터를 10개의 구간으로 나눔, 구간 수를 변경하면 더 세밀한 또는 대략적인 분포를 확인 가능
    - color="skyblue": 막대의 색상을 하늘색으로 설정
    - alpha=0.7: 막대의 투명도를 70%로 설정, 1은 불투명, 0은 완전히 투명
    - edgecolor="black": 막대의 테두리를 검정색으로 설정하여 막대를 구분하기 쉽게 만듬
    - plt.title("Age Distribution in Heart Dataset"): 그래프의 제목을 "Age Distribution in Heart Dataset"으로 설정
    - plt.xlabel("Age"): x축의 이름을 "Age"로 설정, x축은 데이터 값의 범ㅁ위를 나타냄
    - plt,ylabel("Frequency"): y축 이름을 "Frequency"로 설정, y축은 각 구간에 포함된 데이터의 빈도수를 나타냄
    - plt.grid(True): 그래프에 격자를 추가하여 데이터를 더 쉽게 읽을 수 있도록 만듬
    - plt.show(): 설정한 그래프를 화면에 출력
- 실행 결과
    - x축: 나이(age) 값이 구간(bin)으로 나뉘어 표시
    - y축: 각 구간에 포하묀 데이터의 빈도수가 막대 높이로 나타남
    - 그래프: 데이터의 나이 분포를 히스토그램으로 시각화
- 추가 옵션
    - bins 값 변경: 구간 수를 20으로 늘려 데이터의 세밀한 분포를 확인
    - 다른 열 사용: data['chol'](콜레스테롤) 데이터를 히스토그램으로 시각화하여 분포 비교
    - density=True 추가: 밀도 히스토그램으로 변환하여 데이터를 정규화한 분포를 확인

- 박스 그래프(box plot)
    - 박스 그래프란?
        - 박스 그래프는 데이터의 분포와 이상치를 시각적으로 확인하는데 사용, 중앙값, 사분위수 그리고 이상치를 한눈에 보여줌
        - 구성 요소
            - 중앙선: 데이터의 중앙값(Median)을 나타냄
            - 박스: 1사분위(Q1)와 3사분위(Q3) 사이의 범위를 보여줌
            - 수염(Whisker): 박스 빆의 최소값과 최대값을 나타냄
            - 이상치: 수염 밖에 위치한 데이터
        - 활용 사례
            - 그룹별 데이터 비교, ex> 지역별 월 소득 분포
            - 실험 데이터의 분포 및 이상치 분석
            - 고객 리뷰 점수의 분포 및 극단적 평가 확인

```python
import pandas as pd
import matplotlib.pylab as plt

# CSV 데이터 읽기
data = pd.read_csv("datasets/heart.csv")

# 박스 그래프 생성
plt.boxplot(
    data["age"],                                        # 나이 데이터를 시각화
    vert=True,                                          # 그래프를 세로로 표시
    patch_artist=True,                                  # 박스 내부 색상 채우기
    boxprops=dict(facecolor="skyblue", color="black"),  # 박스 스타일
    medianprops=dict(color="red", linewidth=2),         # 중앙값 스타일
    whiskerprops=dict(color="black", linewidth=1.5)     # 상한/하한선 스타일
)

plt.title("Age Distribution in Heart Dataset")          # 그래프 제목 설정
plt.ylabel("Age")                                       # y축 이름 설정
plt.grid(True)                                          # 격자 표시
plt.show()                                              # 그래프 출력
```

- plt.boxplot(): 박스 그래프를 생상하는 함수, 주요 옵션은 아래와 같음
    - data["age"]: 박스 그래프에 사용할 데이터, 여기서는 age 열 데이터 사용

            - y축: 시간에 따른 값의 변화를 나타냄
            - 선: 데이터가 시간에 따라 어떻게 변했는지를 연결하여 보여줌
        - 활용 사계
            - 주식 시장에서 시간별 가격 변화를 시각화
            - 월별 매출 추세 분석
            - 온도 변화 데이터 분석


```python
import pandas as pd
import matplotlib.pylab as plt

# csv 데이터 읽기
data = pd.read_csv("datasets/heart.csv")

# 데이터 정렬 (나이에 따른 정렬)
data = data.sort_values(by="age")

# 선 그래프 생성
plt.plot(
    data["age"],            # x축: 나이
    data["chol"],           # y축: 콜레스테롤 수치
    color="blue",           # 선 색상
    linewidth=2,            # 선 두께
    linestyle="-",          # 선 스타일 (실선)
    marker="o",             # 데이터 점 표시
    markersize=5,           # 점 크기
    label="Cholesterol"     # 범례
)

# 그래프 꾸미기
plt.title("Age vs Cholesterol")     # 그래프 제목
plt.xlabel("Age")                   # x축 이름
plt.ylabel("Cholesterol")           # y축 이름
plt.legend()                        # 범례 표시
plt.grid(True)                      # 격자 표시
plt.show()                          # 그래프 출력             
```

- 막대 그래프(bar plot)
    - 막대 그래프란?
        - 막대 그래프는 서로 다른 그룹 간의 데이터를 비교하는 데 적합, 각 데이터의 크기를 막대의 길이로 표현, 데이터를 직관적으로 비교
    - 구성 요소
        - x축: 비교 대상(ex> 제룸, 지역, 그룹)
        - y축: 해당 항목의 데이터 값
        - 막대: 데이터 값을 시각적으로 나타냄
    - 활용 사례
        - 제품별 매출 데이터 비교
        - 연도별 성장률 분석
        - 고객 선호 제품 유형 비교

```python
import pandas as pd
import matplotlib.pyplot as plt

# csv 데이터 읽기
data = pd.read_csv("datasets/heart.csv")

# 데이터 준비 (카테고리별 평균 콜레스테롤 계산)
age_group = data.groupby("age")["chol"].mean().reset_index()

# 막대 그래프 생성
plt.bar(
    age_group["age"],       # x축: 나이
    age_group["chol"],      # y축: 평균 콜레스테롤
    color="skyblue",        # 막대 색상
    width=0.8,              # 막대 너비
    edgecolor="black"       # 테두리 색상
)

# 그래프 꾸미기
plt.title("Average Cholesterol by Age")     # 그래프 제목
plt.xlabel("Age")                           # x축 이름
plt.ylabel("Average Cholesterol")           # y축 이름
plt.xticks(rotation=45)                     # x축 레이블 회전
plt.grid(axis="y", linestyle="--")          # y축 레이블 회전
plt.show()                                  # 그래프 출력
```

- 파이 그래프(pie plot)
    - 파이 그래프란?
        - 파이 그래프는 데이터의 비율을 시각화하는 데 사용, 전체 데이터에서 각 항목이 차지하는 비중을 직관적으로 보여줌
    - 구성 요소
        - 원형: 데이터의 전체 크기를 나타냄
        - 조각: 각 항목의 비율을 나타냄
    - 활용 사례
        - 시장 점유율 분석, 예산 항목별 비중 비교, 고객 세그먼트 비율 시각화

```python
import matplotlib.pylab as plt

# 데이터 준비
categories = ["Category A", "Category B", "Category C", "Category D"]
values = [30, 20, 35, 15]

# 파이 그래프 생성
plt.pie(
    values,             
    labels=categories,                              # 항목 이름 설정
    autopct="%1.1f%%",                              # 비율 표시 형식
    startangle=90,                                  # 시작 각도 (90도부터 시작)
    explode=(0, 0.1, 0, 0),                         # 특정 항목 강조 (Category B 강조)
    colors=["skyblue", "orange", "green", "red"],   # 섹션 색상
    shadow=True                                     # 그림자 추가
)       

# 그래프 제목 설정
plt.title("Sample Pie Chart")
plt.show()                                          # 그래프 출력
```

- import matplotlib.pyplot as plt: 파이 그래프를 생성하고 시각화 하기 위한 Matplotlib의 Pyplot 모듈을 호출
- categories: 각 항목의 이름, 여기서는 "Category A", "Category B", "Category C", "Category D"로 설정
- values: 각 항목의 값(비율), 값의 총합이 100%로 자동 정규화 되어 표시
- plt.pie(): 파이 그래프를 생성하는 함수, 주요 옵션은 아래와 같음
    - values: 파이 그래프의 각 항목이 차지하는 비율
    - labels=categories: 각 섹션의 이름 설정
    - autopct="%1.1f%%": 각 항목의 비율을 표시
    - startangle=90: 그래프의 시작 각도를 90도로 설정, 이를 통해 그래프를 원하는 방향으로 회전 가능
    - explode=(0, 0.1, 0, 0): 특정 항목을 강조
        - 값이 클수록 원에서 멀리 떨어져 강조, 여기서는 "Category B"를 강조
    - colors: 각 섹션의 색상을 설정, 색상이름(ex> "red") 또는 HTML 색상 코드(ex> "#FF5733")를 사용 가능
    - shadow=True: 그래프에 그림자를 추가하여 입체감 부여
    - plt.title("Sample Pie Chart"): 그래프 상단에 제목을 추가
    - plt.show(): 설정된 파이 그래프를 화면에 출력

| 옵션 | 설명 | 예제 |
|--|--|--|
| labels | 각 항목의 이름을 설정 | labels=["A", "B", "C"] |
| autopct | 각 항목의 비율 표시 형식을 지정 | autopct="%1.1f%%", autopct="%d%%" |
| startangle | 그래프의 시작 각도를 설정 | startangle=90 |
| explode | 특정 항목을 강조하여 원에서 분리 | explode=(0, 0.1, 0, 0) |
| colors | 각 섹션의 색상을 설정 | colors=["red", "blue", "green"] |

```python
# 추가 옵션
# 여러 개의 파이 그래프를 하나의 그림에 배치해 비교

import matplotlib.pyplot as plt


# 데이터 준비
categories = ["Category A", "Category B", "Category C", "Category D"]
values = [30, 20, 35, 15]

# 두 개의 그래프를 위한 서브플롯 생성
fig, axs = plt.subplots(1, 2)

# 첫 번째 파이 그래프
axs[0].pie(
    values, 
    labels=categories,
    autopct="%1.1f%%",
    colors=["skyblue", "orange", "green", "red"]
)

# 두 번째 파이 그래프 (특정 섹션 강조)
axs[1].pie(
    values,
    explode=(0, 0.2, 0, 0),
    labels=categories,
    autopct="%1.1f%%"
)

plt.show()
```

- plt.subplots(1, 2): 하나의 그림(figure) 안에 1행 2열의 서브플롯(작은 그래프 공간)을 생성, 1은 행(Row)의 개수, 2는 열(Column)의 개수
- fig: 전체 그래프 영역(그림)을 관리하는 figure 객체
- axs: 각 서브플롯(작은 그래프 공간)을 제어할 수 있는 객체
- axs[0]: 첫 번째 서브플롯
- axs[1]: 두 번째 서브플롯
